```python
# median : median is the middle value in a sorted data set

# case1:
# odd number of element

[1,2,3,4,5]
# median = 3

# case 2
# even no of element

[1,2,3,4,5,6]

# 3+4/2 = 3.5
# median = 3.5
```

```python
from statistics import median
data = [2,3,4,5,7,9,11]

print("median is ", median(data))

print("---------------------------")
# manually

data = [7,2,3,9,11,4,5]
data.sort()
n = len(data)
if n%2 != 0:    # odd
   median = data[n//2]

else:  # even
 mid1 = data[n//2 -1]
 mid2 = data[n//2]
 median = (mid1+mid2)/2
print("median", median)


```

# **Percentile**

a percentile tells us how a value compares to the rest of the data
it shows what percentage of value are below a certain number


90th percentile - you did better than 90% of the student



```python
import numpy as np
# sample data

scores = [55, 60,70, 80, 85,90, 95]
# calculating percentile
percentile1 = np.percentile(scores , 50)
print(percentile1)

# 50% of the scores are less than or equal to 80.0

percentile2 = np.percentile(scores , 99)
print(percentile2)
```

```python
# outliers:
# outliers are data points that are very different from rest of the data
# outliers can happen because of errors, events or natural varibility

# use case
# detecting outliers help find mistakes in data or special cases invisigation
# fraud detection ,  quality control

```

```python
age = [18, 25, 30, 45, 33, -190, 67,78,89,93,83,250,76, 99, 271]

import matplotlib.pyplot as plt
plt.boxplot(age)
plt.show()
```

```python
q1=np.percentile(age , 25)
q3=np.percentile(age, 75)

iqr = q3-q1
print(q1)
print(q3)
print(iqr)
```

```python
upper_fence  = q3 + 1.5*iqr
lower_fence =  q1-1.5*iqr

print(upper_fence)
print(lower_fence)

# we are setting one bar that any value that is greater than 180.25 and
# any value that is less than -57.75 is going to be outlier
```

```python
for i in age:
  if i> upper_fence or i <lower_fence:
    age.remove(i)


print(age)
```

# **kurtosis**


kurtosis is a way to describe the shape of data distribution.
especially how tall or flat the peak of the graph is compared to
normal curve



Types of kurtosis


1. Leptokurtic(High peak):

The graph has a  very sharp , tall peak


most values are very close to the centre (less spread out)



2. Mesokurtic ( Normal Peak):

the graph has a moderate peak like a normal bell curve

data is evenly spread out around the centre



3. Platykurtic( flat peak):

the graph has a low , flat peak

data is even more spread out and not tightly packed in the centre

```python
data = [15, 16, 16, 17, 17, 17, 17, 17,17, 17, 17, 18, 18,19]
plt.hist(data , bins =5, color ="red", edgecolor ="black")
plt.title("peakedness")
```

# **Skewness**

skewness tell us about asymmetry or lack of symmetry in a data distribution

if data is perfectly symmetric (bell shape ) skewness = 0

if data is more lean towards right or left side then we have either +ve or  -ve

skewness

```python
# types of skewness :


#1. Positive skewness(right skewed):

# the tail on the right side is longer or fatter

# most data points are concentrated on left side
# but there are some very large values that pulling your the mean to right

#  skewness value is greater than 0.


# 2. Negative skewness (left skewed )

# the tail on the left side is flatter

# most data points are concentrated on right side . but there are very few low values
#pulling your mean to the left



# skewness values less than 0



# zero skewness:

# the data is balanced in both sides

# skewness values is 0

# hieght of adult men
```

```python
from scipy.stats import skew

data = [1,2,2,3,4,10,20]
skewness = skew(data)
print(skewness)
```
